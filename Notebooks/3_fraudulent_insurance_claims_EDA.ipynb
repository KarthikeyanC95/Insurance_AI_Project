# **5) Date Features (Year, Month, Day)**


Line Chart: Trends of claims over time.

Heatmap (Day vs. Month): To find seasonality in claims.

```python
# Ensure Month_name is in correct order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Create a new column for sorting purposes
df3['Month_num'] = df3['Month_name'].apply(lambda x: month_order.index(x) + 1)

# Group claims by Year and Month
claims_trend = df3.groupby(['Year', 'Month_name']).size().reset_index(name='Claim_Count')

# Sort by Year and Month number
claims_trend = claims_trend.sort_values(by=['Year', 'Month_name'])

# Create line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x=claims_trend['Month_name'], y=claims_trend['Claim_Count'], hue=claims_trend['Year'], marker='o', palette="tab10")

# Add title and labels
plt.title("Trend of Claims Over Time", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Claims", fontsize=12)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

# **1) Claim_Type vs. Fraud_Label**

Stacked Bar Chart: To check which claim types are associated with fraud.

```python
# Count occurrences of each Claim_Type by Fraud_Label
claim_fraud_counts = df3.groupby(['Claim_Type', 'Fraud_Label']).size().unstack()

# Calculate percentage
claim_fraud_percent = claim_fraud_counts.div(claim_fraud_counts.sum(axis=1), axis=0) * 100

# Plot stacked bar chart with percentages
ax = claim_fraud_percent.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm', edgecolor='black')

# Add percentage labels on bars
for i, bars in enumerate(ax.containers):
    for bar in bars:
        height = bar.get_height()
        if height > 0:  # Only label non-zero bars
            ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2,
                    f'{height:.1f}%', ha='center', va='center', fontsize=10, color='white', fontweight='bold')

# Add title and labels
plt.title("Claim Type vs. Fraud Label (with Percentages)", fontsize=14)
plt.xlabel("Claim Type", fontsize=12)
plt.ylabel("Percentage of Claims", fontsize=12)
plt.xticks(rotation=45)

# Add legend
plt.legend(title="Fraud Label", labels=["Non-Fraud", "Fraud"])

# Show plot
plt.show()
```

# **Insights from the Claim Type vs. Fraud Label Chart:**

**Dominance of Non-Fraudulent Claims:**

For all three claim types **(Home Damage, Medical, and Vehicle)**, **the number of non-fraudulent claims significantly outweighs the number of fraudulent claims**. This is expected in most insurance scenarios.


**Relatively Consistent Fraud Proportion:**

While the absolute numbers vary, the proportion of fraudulent claims seems relatively consistent across the three claim types. **Medical claims might have a slightly higher proportion of fraud**, but further statistical testing would be needed to confirm this.


**Potential for Further Investigation:**

The chart highlights areas where further investigation might be warranted. For instance:


**Medical Claims:**

 The slightly higher proportion of potential fraud in medical claims could be explored further to identify specific patterns or vulnerabilities.


**High-Value Claims:**

While not shown in this chart, analyzing fraud specifically within **high-value claims could be insightful**, as these would have a **greater financial impact.**

# **2) Year, Month_name vs. Fraud_Label**

Bar Chart: To analyze fraud occurrences per month.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure Month_name follows the correct order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Get unique years
unique_years = sorted(df3['Year'].unique())

# Set figure size dynamically
plt.figure(figsize=(12, len(unique_years) * 6))

# Loop through each year to create separate plots
for i, year in enumerate(unique_years, 1):
    plt.subplot(len(unique_years), 1, i)  # Create a subplot for each year

    # Filter data for the specific year
    yearly_data = df3[df3['Year'] == year]

    # Count occurrences of each Month_name by Fraud_Label
    month_fraud_counts = yearly_data.groupby(['Month_name', 'Fraud_Label']).size().unstack()

    # Reorder months and fill missing values with 0
    month_fraud_counts = month_fraud_counts.reindex(month_order, fill_value=0)

    # Convert counts to percentages
    month_fraud_percent = month_fraud_counts.div(month_fraud_counts.sum(axis=1), axis=0) * 100

    # Plot stacked bar chart with percentages
    ax = month_fraud_percent.plot(kind='bar', stacked=True, colormap='coolwarm', edgecolor='black', ax=plt.gca())

    # Add percentage labels on bars
    for bars in ax.containers:
        for bar in bars:
            height = bar.get_height()
            if height > 0:  # Only label non-zero bars
                ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + height / 2,
                        f'{height:.1f}%', ha='center', va='center', fontsize=10, color='white', fontweight='bold')

    # Add title and labels
    plt.title(f"Fraud Cases by Month - Year {year} (Percentage)", fontsize=14)
    plt.xlabel("Month", fontsize=12)
    plt.ylabel("Percentage of Claims", fontsize=12)
    plt.xticks(rotation=45)

    # Add legend only for the first plot to avoid repetition
    if i == 1:
        plt.legend(title="Fraud Label", labels=["Non-Fraud", "Fraud"])
    else:
        plt.legend().remove()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

# **3) Claim_Amount vs. Fraud_Label**

Box Plot: To check if fraudulent claims have different amount distributions.

```python
# Calculate percentage of each Fraud_Label category
fraud_counts = df3['Fraud_Label'].value_counts(normalize=True) * 100  # Convert to percentage

# Set figure size
plt.figure(figsize=(10, 6))

# Create box plot
ax = sns.boxplot(x='Fraud_Label', y='Claim_Amount', data=df3, palette='coolwarm')

# Add percentage labels on top of each box
for i, label in enumerate(fraud_counts.index):
    percentage_label = f"{fraud_counts[label]:.1f}%"  # Format percentage
    ax.text(i, df3['Claim_Amount'].max(), percentage_label,
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Add title and labels
plt.title("Claim Amount Distribution by Fraud Label (With Percentage)", fontsize=14)
plt.xlabel("Fraud Label", fontsize=12)
plt.ylabel("Claim Amount", fontsize=12)
plt.xticks(ticks=[0, 1], labels=["Non-Fraud", "Fraud"])

# Show the plot
plt.show()
```

Fraudulent claims tend to be for larger amounts of money compared to legitimate claims.

The amounts of fraudulent claims are also more unpredictable and vary more widely.

While claim amount can be a clue, it shouldn't be the only factor used to detect fraud, as there's some overlap with legitimate claim amounts.

**Risk Assessment:**

These findings can inform risk assessment strategies. **Claims for larger amounts might warrant closer** scrutiny for potential fraud.

**Fraud Detection Models:**

Claim amount could be a useful feature **in fraud detection models, especially when combined with other relevant variables**.

**Resource Allocation:**

**Investigating potentially fraudulent claims, particularly those for larger amounts**, might be a more efficient use of resources.

# **4) Claim_Amount vs. Suspicious_Flags**

Scatter Plot: To analyze correlation between flagged claims and amount.

```python
# Calculate percentage of each Suspicious_Flags category
flag_counts = df3['Suspicious_Flags'].value_counts(normalize=True) * 100  # Convert to percentage

# Set figure size
plt.figure(figsize=(10, 6))

# Create scatter plot
ax = sns.scatterplot(x='Suspicious_Flags', y='Claim_Amount', data=df3, alpha=0.6, hue='Suspicious_Flags', palette='coolwarm')

# Add percentage labels near the x-axis
for i, label in enumerate(flag_counts.index):
    percentage_label = f"{flag_counts[label]:.1f}%"  # Format percentage
    ax.text(i, df3['Claim_Amount'].max(), percentage_label,
            ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Add title and labels
plt.title("Scatter Plot: Claim Amount vs. Suspicious Flags (With Percentage)", fontsize=14)
plt.xlabel("Suspicious Flags", fontsize=12)
plt.ylabel("Claim Amount", fontsize=12)

# Show the plot
plt.show()
```

# **False (0) -> Genuine**
