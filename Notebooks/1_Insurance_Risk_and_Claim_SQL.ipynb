# **Connect SQL**

```python
import sqlite3
import pandas as pd

# File name of the uploaded CSV
csv_filename = "/content/drive/MyDrive/Captsone project/Data/df1-insurance_risk_claim_dataset.csv"  # Replace this with the uploaded file name

# Load CSV into a pandas DataFrame
df = pd.read_csv(csv_filename)

# Connect to SQLite database (or create a new one)
conn = sqlite3.connect("example.db")
cursor = conn.cursor()

# Write DataFrame to SQLite table
table_name = "df1"  # Specify your table name
df.to_sql(table_name, conn, if_exists="replace", index=False)

print(f"Table '{table_name}' created in SQLite database.")
```

# **1) Retrieve all records**

**Write a query to select all records from the dataset.**

```python
query = f"SELECT * from df1;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **2) Filter by Age**

**Write a query to find all policies where the Customer_Age is greater than 60.**

```python
query = f"SELECT Policy_ID,Customer_Age,Gender,Policy_Type,Annual_Income,Vehicle_Age_Property_Age,Claim_History,Fraudulent_Claim,Premium_Amount,Claim_Amount,Risk_Score from df1 where Customer_Age>=60 order by Customer_Age ASC;"
result = pd.read_sql_query(query, conn)

# Display the results
result
# Display the results
result
```

# **3) Count Policies by Gender**

**Write a query to count the number of policies for each Gender**

```python
query = f"SELECT Gender,count(Policy_ID) from df1 GROUP BY Gender;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **4) Average Premium Amount**

**Write a query to calculate the average Premium_Amount.**

# **5) Identify High-Risk Policies**

# **6) Find Total Claim Amount**

**Write a query to sum up the total Claim_Amount for all policies**

```python
query = f"SELECT Policy_Type,sum(Claim_Amount) as Total_Claims FROM df1 Group by Policy_Type Order by Total_Claims DESC"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **7) Detect Fraudulent Claims**

**Write a query to list policies where Fraudulent_Claim is greater than 1**

**1 = Fraud, 0 = Genuine**

```python
query = f"SELECT Policy_Type,COUNT(Fraudulent_Claim) as Fraud_Claims FROM df1 Where Fraudulent_Claim=1 Group by Policy_Type Order by Fraud_Claims DESC "
result = pd.read_sql_query(query, conn)

# Display the results
result
```

```python
query = f"SELECT Policy_Type,COUNT(Fraudulent_Claim) as Genune_Claims FROM df1 Where Fraudulent_Claim=0 Group by Policy_Type ORDER by Genune_Claims DESC"
result = pd.read_sql_query(query, conn)

# Display the results
result
```
