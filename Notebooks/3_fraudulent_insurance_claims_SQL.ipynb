```python
import sqlite3
import pandas as pd

# File name of the uploaded CSV
csv_filename = "/content/drive/MyDrive/Captsone project/Data/df3-date-modified.csv"  # Replace this with the uploaded file name

# Load CSV into a pandas DataFrame
df = pd.read_csv(csv_filename)

# Connect to SQLite database (or create a new one)
conn = sqlite3.connect("example.db")
cursor = conn.cursor()

# Write DataFrame to SQLite table
table_name = "df3"  # Specify your table name
df.to_sql(table_name, conn, if_exists="replace", index=False)

print(f"Table '{table_name}' created in SQLite database.")
```

# **I) Basic Queries**

# **1) Retrieve all claim details**

```python
query = f"SELECT * from df3;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **2) Find the total number of claims**

```python
query = f"SELECT count(Claim_ID) from df3"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **3) Find the number of fraud vs. non-fraud claims**

```python
query = f"SELECT DISTINCT Fraud_Label,count(Fraud_Label) from df3 GROUP BY Fraud_Label"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **Fraud_Label -> ( 0 - Genuine, 1 - Fraud )**

# **Claim Amount Analysis**

#**4) Find the average claim amount per claim type**

```python
query="Select Claim_Type,avg(Claim_Amount) as AVG_CLAIM from df3 GROUP BY Claim_Type ORDER by AVG_CLAIM DESC"
result=pd.read_sql_query(query,conn)

result
```

**Medical	 as a Highest Average Claim Amount**

**Vehicle as Medium Average Claim**

**Home Damage is Least Average Claim**

# **5) Find the highest and lowest claim amounts**

```python
query="Select max(Claim_Amount) as Highest_Claim_Amount,min(Claim_Amount) as Lowest_Claim_Amount from df3"

result=pd.read_sql_query(query,conn)
result
```

# **Highest and Lowest claim amounts based on Claim Type**

```python
query="Select Claim_Type,max(Claim_Amount) as Highest_Claim_Amount,min(Claim_Amount) as Lowest_Claim_Amount from df3 GROUP BY Claim_Type;"

result=pd.read_sql_query(query,conn)
result
```

# **6) Find claims with an amount greater than $25,000**

```python
query="Select * From df3 where Claim_Amount > 25000;"

result=pd.read_sql_query(query,conn)
result
```

# **Count of claims with an amount greater than $25,000 on Each Claim Types**

```python
query="Select Claim_Type,count(Claim_Amount) as Count_Claim_Amount From df3 WHERE Claim_Amount >= 25000 GROUP BY Claim_Type"

result=pd.read_sql_query(query,conn)
result
```

# **Fraud Detection & Suspicious Flags**


# **7) Find the total number of flagged claims**

```python
query="Select count(Suspicious_Flags) as Flagged_Claims From df3 where Suspicious_Flags=1;"

result=pd.read_sql_query(query,conn)
result
```

# **‚úÖ True (1) ‚Üí The claim is flagged as suspicious (potential fraud or anomaly detected).**

# **‚ùå False (0) ‚Üí The claim is not suspicious (no immediate fraud indicators).**

# **8) Find the most common claim type for fraudulent claims**

# **Genuine = 1 & Fraud = 0**

```python
query="Select Claim_Type,Count(Fraud_Label) as Fraudlent_Claims From df3 where Fraud_Label=1 GROUP BY Claim_Type ORDER BY Fraudlent_Claims DESC;"

result=pd.read_sql_query(query,conn)
result
result = pd.read_sql_query(query,conn)

result
```

# **Vehicle Claim Type Most Fraudlent Happens**

# **9) Find the percentage of fraud claims per claim type**

```python
query="Select Claim_Type,count(CASE when Fraud_Label=1 Then 1 END)*100/count(*) as Fraud_Percentage From df3 Group By Claim_Type ORDER BY Fraud_Percentage DESC;"

result=pd.read_sql_query(query,conn)
result
```

# **Time-Based Analysis**


# **10) Find the number of claims per year**

```python
query="Select Year,count(Claim_ID) as Claims From df3 GROUP BY Year"

result=pd.read_sql_query(query,conn)
result
```

# **2023 has Highest Claims**

# **11) Find the number of Fraud Claims per Year**

**0 = Genuine, 1 = Fraud**

```python
query="Select Year,count(Claim_ID) as Claims From df3 where Fraud_Label=1 GROUP BY Year"

result=pd.read_sql_query(query,conn)
result
```

# **2023 has highest Fraud Claims**

# **12) Find the month with the highest number of claims**

```python
query="Select month_name,Year,count(Claim_Amount) as Total_claims,sum(Claim_Amount) as Sum_Claims From df3 GROUP BY month_name order by Total_claims DESC"

result=pd.read_sql_query(query,conn)
result
```

# **Policyholder Insights**

# **13) Find the total number of unique policyholders**

```python
query="Select DISTINCT Policyholder_ID,count(Policyholder_ID)  From df3 Group By Policyholder_ID"  #Policyholder_ID

result=pd.read_sql_query(query,conn)
result
```

# **14) Find the policyholder with the highest total claim amount**

```python
query="Select Policyholder_ID,Count(Claim_Amount),max(Claim_Amount) as Max_Claims,sum(Claim_Amount) as Total_Amount From df3 Group By Policyholder_ID ORDER BY Total_Amount DESC"  #Policyholder_ID

result=pd.read_sql_query(query,conn)
result
```
