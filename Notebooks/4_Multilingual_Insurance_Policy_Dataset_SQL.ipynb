```python
import sqlite3
import pandas as pd

# File name of the uploaded CSV
csv_filename = "/content/drive/MyDrive/Captsone project/Data/df4-insurance_multilingual_policy_document_dataset.csv"  # Replace this with the uploaded file name

# Load CSV into a pandas DataFrame
df = pd.read_csv(csv_filename)

# Connect to SQLite database (or create a new one)
conn = sqlite3.connect("example.db")
cursor = conn.cursor()

# Write DataFrame to SQLite table
table_name = "df4"  # Specify your table name
df.to_sql(table_name, conn, if_exists="replace", index=False)

print(f"Table '{table_name}' created in SQLite database.")
```

# **I) Basic Retrieval and Filtering**

# **1) Retrieve Policy and Summary**

```python
query = f"SELECT * from df4;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **2) Find Policies Containing a Keyword in the Full Text**

```python
query = f"SELECT * from df4 WHERE policy_text LIKE '%tend%';"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **Summary and Aggregation**

# **3) Count the Total Number of Policies**

```python
query = f"SELECT count(policy_ID) as total_policies from df4 WHERE policy_text LIKE '%tend%';"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **4) Find the Average Length of the Summarized Text**

```python
query = f"SELECT AVG(LENGTH(summarized_text)) AS avg_summary_length from df4;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **Text Analysis and Comparison**

# **5) Find Policies Where the Summary is Empty or NULL**

```python
query = f"SELECT * from df4 WHERE summarized_text IS NULL OR summarized_text = '';"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **6) Retrieve Policies Where the Summary is Less Than 10% of the Full Text Length**

```python
query = f"SELECT policy_id, LENGTH(policy_text) AS full_length, LENGTH(summarized_text) AS summary_length from df4 WHERE LENGTH(summarized_text) < (LENGTH(policy_text)*0.1);"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **Data Manipulation**

# **7) Find the Summarized Text for a Specific Policy**

```python
query = f"SELECT summarized_text from df4 WHERE policy_id like '%205%';"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **8) Find Policies Where the Summary Mentions a Term But the Full Text Does Not**

```python
query = f"SELECT policy_id from df4 WHERE summarized_text LIKE '%tend%' AND policy_text NOT LIKE '%tend%';"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **Advanced Ordering and Ranking**

# **9) Retrieve the Top 5 Longest Policy Texts**

```python
query = f"SELECT policy_id, LENGTH(policy_text) AS text_length from df4 ORDER BY text_length DESC LIMIT 5;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```

# **10) Group and Count Policies Based on a Keyword in the Summary**

```python
query = f"SELECT SUBSTRING(summarized_text, 1, INSTR(summarized_text, ' ')) AS category, COUNT(policy_id) AS policy_count from df4 WHERE summarized_text LIKE '% %' GROUP BY category ORDER BY policy_count DESC;"
result = pd.read_sql_query(query, conn)

# Display the results
result
```
