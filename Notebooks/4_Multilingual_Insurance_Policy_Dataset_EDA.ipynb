# **Description: Collection of insurance policies in multiple languages for NLP-based translation and summarization.**

```python
import pandas as pd
df4=pd.read_csv(r'/content/drive/MyDrive/Captsone project/Data/df4-insurance_multilingual_policy_document_dataset.csv')
df4.head()
```

# **Create wordcloud for TEXT column for better Understanding**

# **1) Policy_Text**

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load or create the DataFrame (df1 should already exist)
# If df1 is not created yet, you can generate it using the previous script

# Select only the text-based columns
text_columns = ['Policy_Text']

# Combine text data from all selected columns
# The change is in the following line: iterating over values in the series to create a list of strings
text_data = " ".join([str(value).lower() for col in text_columns for value in df4[col].dropna()])

# Generate the Word Cloud
wordcloud = WordCloud(width=1000, height=500, background_color='white', colormap='viridis').generate(text_data)

# Display the Word Cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")  # Hide the axes
plt.title("Word Cloud of Textual Data from Insurance Multilingual Policy Document Dataset", fontsize=14)
plt.show()
```
